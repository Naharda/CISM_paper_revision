```python
import pandas as pd

```

```python
df = pd.read_csv("Melanona_MIBI_TOF_18_7_2024.csv")
```

```python
df.head()
```

```python
len(df['pred'].unique())
```

```python
wanted_cols = ['pred', 'meta_pred']

df[wanted_cols].drop_duplicates()
```

```python
translator = {}
for cell_type in df['pred'].unique():
    
    cell_type = cell_type.strip().lower()

    if 'mac' in cell_type or 'mono_cd14_dr' == cell_type:
        translator[cell_type] = 'mac'
        
    elif 'collagen' in cell_type or 'sma' in cell_type:
        translator[cell_type] = 'stroma'
    
    elif 'cd45ra+mhc2+' in cell_type:
        translator[cell_type] = 'b cell'
    
    elif cell_type in ['cd14_cd11c_dcs','cd11_cd11c_dcsign_dcs']:
        translator[cell_type] = 'DC'
        
    elif cell_type == 'tfh%cd4 t cell' or cell_type == 'cd4 apc':
        translator[cell_type] = 'cd4 t cell'


    elif cell_type == 'cd4 treg%cd4 t cell':
        translator[cell_type] = 'Treg'
    
    elif cell_type == 'follicular_germinal_b_cell':
        translator[cell_type] = 'gc'
        
    else:
        translator[cell_type] = cell_type
        
translator
```

```python
df['pred_mapped'] = df['pred'].apply(lambda x: translator[x.strip().lower()])
df
```

df.to_csv('mapped_cell_types.csv', index=False)
